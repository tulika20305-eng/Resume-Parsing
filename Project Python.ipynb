# I/P & O/P CREATION

```python
ip = resume2.drop("Experience Level",axis=1)
op = resume2["Experience Level"]
```

# TRAIN TEST SPLIT

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(ip,op,test_size=0.2)
```

# STANDARD SCALER

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

# MACHINE LEARNING MODEL

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train,y_train)
```

# PREDICTION

```python
ypred = lr.predict(x_test)
ypred
```

# ACCURACY

```python
from sklearn.metrics import accuracy_score,recall_score,f1_score,precision_score
acc = accuracy_score(ypred,y_test)
recall = recall_score(ypred,y_test,average="weighted")
precision = precision_score(ypred,y_test,average="weighted")
f1 = f1_score(ypred,y_test,average="weighted")
print("Accuracy:",acc)
print("Recall:",recall)
print("Precision:",precision)
print("f1_score:",f1)
```

```python
df = pd.DataFrame({"prediction":ypred,"Actual data":y_test})
sns.lmplot(df,x="prediction",y="Actual data")
plt.show()
```

```python
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
cm = confusion_matrix(ypred,y_test)
cm
```

```python
cmd = ConfusionMatrixDisplay(cm)
cmd.plot()
plt.show()
```

```python
# 2ND MODEL
from sklearn.svm import SVC
svm = SVC(kernel="rbf",C=1000,gamma=0.01)
svm.fit(x_train,y_train)
```
